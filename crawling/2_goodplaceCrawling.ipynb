# 무신사크롤링

```python
import re
```

```python
for i in range(1, 11):
    print('Page ============================================', i)
    url = f'https://www.musinsa.com/categories/item/001?d_cat_cd=001&brand=&list_kind=small&sort=pop_category&sub_sort=&page={i}&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&kids=&color=&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure='
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    for item in soup.select('ul#searchList > li'):
        print('상품명 : ', item.select_one('.article_info p.list_info a').attrs['title'])
        price = '';
        if len(re.findall('[0-9,]*원', item.select_one('.article_info p.price').text )) == 1:
            price = re.findall('[0-9,]*원', item.select_one('.article_info p.price').text)[0]
        else:
            price = re.findall('[0-9,]*원', item.select_one('.article_info p.price').text)[1]
        print('가격 : ', price)
        print('')
```

# 식신크롤링

```python
import pandas as pd
from urllib.parse import quote
```

```python
# https://www.siksinhot.com/search?keywords=%EC%8B%A0%EB%85%BC%ED%98%84%EC%97%AD
site = 'https://www.siksinhot.com/ko/'
url = f'{site}/search?keywords={quote("신논현역")}'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

## 필요한 데이터 찾기

```python
lis = soup.find_all('li')
len(lis)
```

```python
lis = soup.select_one('.localFood_list').find_all('li')
len(lis)
```

```python
line = []
for li in lis:
    store = li.select_one('.textBox > h2').get_text()
    try:
        score = float(li.select_one('.score').get_text())
    except:
        score = 0.0
    place = li.select_one('.cate').get_text().split('/')[0].strip()
    menu = li.select_one('.cate').get_text().split('/')[1].strip()
    line.append({'상호' : store, '평점' : score, '위치' : place, '메뉴':menu})
```

```python
df = pd.DataFrame(line)
df
```
